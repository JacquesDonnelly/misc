```python
import re

from bs4 import BeautifulSoup
import requests
```

```python
url = "https://quizlet.com/_7rddz8?x=1jqt&i=2lt25u"
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
r = requests.get(url, headers={'User-Agent': user_agent})
r.raise_for_status()
soup = BeautifulSoup(r.text)
```

### Extract Expected Number of Terms

```python
num_term_h4 = soup.find("h4").text
match = re.search("terms in this set \((.*)\)", num_term_h4, re.IGNORECASE)
num_terms = int(match.group(1))
num_terms
```

### Extract Terms

```python
termsList = soup.find_all(attrs={"class": "SetPageTerm-inner"})
term = termsList[0]

cards = []
for term in termsList:
    card = [
        link.get_text(separator="<br>").replace("<br>", "\n")
        for link in term.find_all("a")
    ]
    cards.append(card)

assert len(cards) == num_terms    

for card in cards:
    print(card[0])
    print("\n")
    print(card[1])
    print("\n")
```
