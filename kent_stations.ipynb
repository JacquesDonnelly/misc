```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text)
```

```python
print(soup.prettify())
```

```python
table = soup.find("table", attrs={"class": 'wikitable sortable'})
```

```python
table
```

```python
table_body = table.find('tbody')
```

```python
table_body
```

```python
rows = table_body.find_all('tr')
```

```python
row = rows[1]
```

```python
row_name = row.find('th')
```

```python
row_name.text.strip()
```

```python
stations = [row.find('th').text.strip() for row in rows]
```

```python
s = pd.Series(stations[1:])
```

```python
df = s.str.extract("(.*) \((.*)\)")
```

```python
df.to_csv("kent_stations.csv")
```

```python
station = df[1].iloc[2]
```

```python
url = f"http://ojp.nationalrail.co.uk/service/timesandfares/{station}/CST/060120/0830/arr"
```
